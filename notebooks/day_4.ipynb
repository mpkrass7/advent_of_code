```python
def read_file(file_path):
    f= open(file_path, 'r')
    final_list=[]
    for line in f:
        final_list.append(line)
    return final_list

def sep_passports(text):
    final_list=[]
    mid_dict = {}
    for line in text: 
        if line == '\n':
            final_list.append(mid_dict)
            mid_dict = {}
        else:
            line_form = line.replace('\n', '')
            line_form = line_form.split(' ')
            for item in line_form:
                x = item.split(':')
                mid_dict[x[0]] = x[1]
    return final_list

def count_valid_passports(text_dicts):
    count = 0
    req_entries = set(['eyr','iyr','byr','ecl','pid','hcl','hgt'])
    for item in text_dicts:
        start = req_entries.difference(item)
        if len(start) == 0:
            count+=1
    return count

passport_text = read_file('day_4/passports.txt')
list_of_dicts = sep_passports(passport_text)
count_valid_passports(list_of_dicts)
```
